### The duplicates are mostly for complaints about rodents. People may call at the same time to complain and we can drop them.

```python
#Drop the duplicates
UniqueData = data.drop_duplicates(['Unique Key'])
print (UniqueData.shape)
```

```python
#What columns have missing vales?
UniqueData.isnull().sum()
```

```python
com_grp = UniqueData[['Community Board','Borough','Agency']].groupby(["Community Board",'Borough'])
com_dists = com_grp.agg('count')
com_dists.head()
```

```python
# Generate marker for unplausible Community Districts
# How do these districts look like? 
# Code borrowed from Scott Smith's HW8

UniqueData["valid_board"] = np.ones(len(UniqueData.Borough), np.float)
for bname in com_dists.itertuples():
    if re.match(r"[01]\d", bname[0][0]) == None:
        UniqueData.valid_board.loc[UniqueData['Community Board'] == bname[0][0]] = np.nan
UniqueData.valid_board.isnull().sum()
```

```python
# Drop the marked districts
UniqueData = UniqueData[~UniqueData['valid_board'].isnull()]
```

```python
plausibleCB = UniqueData['Community Board'].unique()
plausibleCB.sort()
print (plausibleCB)
```

```python
#Create new column to count the complaints
UniqueData['Complaints'] = np.zeros(len(UniqueData['valid_board']))
com_grp = UniqueData[['Community Board', 'Borough',
                    'Complaints']].groupby(['Community Board', 'Borough'])
comp_by_board = com_grp.agg({'Complaints' : 'count'})
```

```python
#Sort by Number of Complaints
comp_by_board.sort('Complaints')
```

```python
#get demographic data
demog = pd.read_csv('http://cosmo.nyu.edu/~fb55/PUI2016/data/Final_Demographics.csv')
demog.head()
```

```python
#How many community districts are in the file?
print ("There are " + str(len(demog['cd_id'])) + " community districts represented in the file.")
```

```python
#Find out some information by District

# Set some variables to refer to column names in df_pop
# Use these variables to refer to columns instead of the full names
Popn = "Total Population"
Dense = "Population Density (per sq. mile)"
PctEnglish = "% Population 5 Years And Over: Speak Only English"
PctEngWell = '% Population 5 Years And Over: Spanish or Spanish Creole: Speak English "very Well"'
PctEngNotWell = '% Population 5 Years And Over: Spanish or Spanish Creole: Speak English Less Than "very Well"'
PctBach = "% Population 25 Years and over: Bachelor's degree or more"
PctMaster = "% Population 25 Years and over: Master's degree or more"
PctProf = "% Population 25 Years and over: Professional school degree or more"
PctDoctorate = "% Population 25 Years and over: Doctorate degree.1"
Bach = "Population 25 Years and over: Bachelor's degree or more"
Master = "Population 25 Years and over: Master's degree or more"
Prof = "Population 25 Years and over: Professional school degree or more"
Doctorate = "Population 25 Years and over: Doctorate degree.1"

```

```python
# Manipulate data to get some information on demographics by Community District. 
# Who might call 311?
# Two factors. One, the likelihood of reasons in their neighborhood to call 311.
# Two, propensity to react to one of those reasons.

# Preponderance of reasons would come with density, low income, and not speaking English well
# Likeliness of responding to such reasons would come from education
demog['Low Income'] = (demog["Households: Less than $10,000"] +
                        demog["Households: $10,000 to $14,999"] +
                        demog["Households: $15,000 to $19,999"] +
                        demog["Households: $20,000 to $24,999"] +
                        demog["Households: $25,000 to $29,999"])
demog['Pct Low Income'] = (demog["% Households: Less than $10,000"] +
                        demog["% Households: $10,000 to $14,999"] +
                        demog["% Households: $15,000 to $19,999"] +
                        demog["% Households: $20,000 to $24,999"] +
                        demog["% Households: $25,000 to $29,999"])
demog_income = demog[["FIPS", "cd_id", Popn, Dense, PctEnglish, PctEngWell, PctEngNotWell,
                      Bach, Master, Prof, Doctorate, "Low Income", "Pct Low Income"]]

demog_income
```

```python
#rename columns so they match the community district in the 311 dataset
demog['cd_id'] = demog['cd_id'].map(lambda x: x.replace('BX', 'BRONX '))
demog['cd_id'] = demog['cd_id'].map(lambda x: x.replace('MN', 'MANHATTAN '))
demog['cd_id'] = demog['cd_id'].map(lambda x: x.replace('BK', 'BROOKLYN '))
demog['cd_id'] = demog['cd_id'].map(lambda x: x.replace('QN', 'QUEENS '))
demog['cd_id'] = demog['cd_id'].map(lambda x: x.replace('SI', 'STATEN_ISLAND '))
```

```python
#clean the error in Manhattan 11
demog['cd_id'] = demog['cd_id'].map(lambda x: x.replace('11111', '11'))
```

```python
demog['cd_id']
```

```python
#switch the order of the community district number and borough to align with the 311 dataset
for i in range (0,len(demog)):
    demog['cd_id'][i] = (demog['cd_id'].str.split(' '))[i][1] + ' ' + (demog['cd_id'].str.split(' '))[i][0]
```

```python
demog['cd_id']
```

```python
#Rename Staten Island to match the 311 dataset
demog['cd_id'] = demog['cd_id'].map(lambda x: x.replace('STATEN_ISLAND', 'STATEN ISLAND '))
```
